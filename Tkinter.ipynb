```python
from tkinter import *
from PIL import ImageTk,Image

root=Tk()
root.title("IMAGES")
root.iconbitmap('teemo.ico')

my_image1=ImageTk.PhotoImage(Image.open("building.jpg"))
my_image2=ImageTk.PhotoImage(Image.open("astronaut.jpg"))
my_image3=ImageTk.PhotoImage(Image.open("Teemo1.jpg"))

image_list=[my_image1,my_image2,my_image3]

status=Label(root,text="Image 1 of "+str(len(image_list)),bd=1,relief=SUNKEN)

def forward(n):
    global my_label
    global button_forward
    global button_back
    
    my_label.grid_forget()
    my_label=Label(image=image_list[n-1])
    
    button_back=Button(root,text='<<',command=lambda:back(n-1))
    button_next=Button(root,text='>>',command=lambda:forward(n+1))

    if(n==len(image_list)):
        button_next=Button(root,text='>>',state=DISABLED)
    
    my_label.grid(row=0,column=0,columnspan=3)
    button_back.grid(row=1,column=0)
    button_next.grid(row=1,column=2)
    
    status=Label(root,text="Image " +str(n)+" of "+str(len(image_list)),bd=1,relief=SUNKEN)
    status.grid(row=2,column=0,columnspan=3,sticky=W+E)
    
def back(n):
    global my_label
    global button_forward
    global button_back
    
    my_label.grid_forget()
    my_label=Label(image=image_list[n-1])
    
    button_back=Button(root,text='<<',command=lambda:back(n-1))
    button_next=Button(root,text='>>',command=lambda:forward(n+1))
    
    if(n==1):
        button_back=Button(root,text='<<',state=DISABLED)
    
    my_label.grid(row=0,column=0,columnspan=3)
    button_back.grid(row=1,column=0)
    button_next.grid(row=1,column=2)
    
    t="Image " +str(n)+" of "+str(len(image_list))
    print(type(t))
    status=Label(root,text=t,bd=1,relief=SUNKEN,anchor=E)
    status.grid(row=2,column=0,columnspan=3,sticky=W+E)
    

my_label=Label(image=image_list[0])
my_label.grid(row=0,column=0,columnspan=3)

button_back=Button(root,text='<<',command=back,state=DISABLED)
button_exit=Button(root,text='Exit Program',command=root.quit)
button_next=Button(root,text='>>',command=lambda:forward(2))

button_back.grid(row=1,column=0)
button_exit.grid(row=1,column=1)
button_next.grid(row=1,column=2,pady=10)

status.grid(row=2,column=0,columnspan=3,sticky=W+E)

root.mainloop()
```

```python
from tkinter import *
from PIL import ImageTk,Image
import sqlite3

root=Tk()
root.title("DataBase")
root.iconbitmap("teemo.ico")
root.geometry("500x500")

connection=sqlite3.connect("address_book.db")

cursor=connection.cursor()



f_name=Entry(root,width=30)
f_name.grid(row=0,column=1,padx=20)

l_name=Entry(root,width=30)
l_name.grid(row=1,column=1)

city=Entry(root,width=30)
city.grid(row=2,column=1)

zip_code=Entry(root,width=30)
zip_code.grid(row=3,column=1)

delete_box=Entry(root,width=5)
delete_box.grid(row=9,column=1)


#
f_name_label=Label(root,text="First Name:")
f_name_label.grid(row=0,column=0)

l_name_label=Label(root,text="Last Name:")
l_name_label.grid(row=1,column=0)

city_label=Label(root,text="City:")
city_label.grid(row=2,column=0)

zip_code_label=Label(root,text="Zip code:")
zip_code_label.grid(row=3,column=0)

delete_box_label=Label(root,text="ID number:").grid(row=9,column=0)
#

def submit():
    connection=sqlite3.connect("address_book.db")
    cursor=connection.cursor()
    
    cursor.execute("INSERT INTO addresses VALUES (:f_name,:l_name,:city,:zip_code)",
              {
                  'f_name':f_name.get(),
                  'l_name':l_name.get(),
                  'city':city.get(),
                  'zip_code':zip_code.get()                  
              })
    
    connection.commit()
    connection.close()
    
    
    
    
    
    
    f_name.delete(0,END)
    l_name.delete(0,END)
    city.delete(0,END)
    zip_code.delete(0,END)
    
        

submit_button=Button(root,text="Add Record to DataBase",command=submit)
submit_button.grid(row=5,column=0,columnspan=2,padx=10,ipadx=100)

#Creating a query button
def query():
    connection=sqlite3.connect("address_book.db")
    cursor=connection.cursor()
    
    cursor.execute("SELECT *,oid FROM addresses")
    records = cursor.fetchall( )
    
    print_records=""
    for i in records:
        print_records+=str(i)+"\n"
    query_label=Label(root,text=print_records).grid(row=8,column=0)
    
    connection.commit()
    connection.close()

query_button=Button(root,text="Show Records",command=query)
query_button.grid(row=7,column=0,columnspan=2,padx=10,pady=10,ipadx=137)

def delete(number):
    connection=sqlite3.connect("address_book.db")
    cursor=connection.cursor()
    
    cursor.execute("DELETE from addresses WHERE oid="+str(number))
        
    
    connection.commit()
    connection.close()
    

delete_button=Button(root,text="Delete Record",command=lambda:delete(delete_box.get()))
delete_button.grid(row=10,column=0,columnspan=2,padx=10,pady=10,ipadx=137)


connection.commit()

connection.close()

root.mainloop()
```
